# Import Required Libraries
Import the necessary libraries, including IFrame, HTML, display from IPython.display and os.

```python
from IPython.display import IFrame, HTML, display
import os
```

# Define File Path
Define the file path to the HTML file you want to display.

```python
# Define the file path to the HTML file you want to display
file_path = "../data/SRR5088813/SRR5088813_2_fastqc.html"
```

# Check File Existence
Check if the file exists using os.path.exists().

```python
# Check if the file exists first
if os.path.exists(file_path):
    print(f"File found: {file_path}")
else:
    print(f"File not found: {file_path}")
    print("Please check if the path is correct and the file exists.")
```

# Display HTML File
Use the display() function with IFrame to display the HTML file if it exists.

```python
# Display HTML File
if os.path.exists(file_path):
    # Display the HTML file using IFrame
    display(IFrame(src=file_path, width=200, height=200, scrolling=True))
else:
    print(f"File not found: {file_path}")
    print("Please check if the path is correct and the file exists.")
```

```python
from IPython.display import HTML, display
import os

# Define the file path
file_path = "../data/SRR5088813/SRR5088813_1_fastqc.html"

# Check if the file exists first
if os.path.exists(file_path):
    # Read the HTML content
    with open(file_path, 'r') as file:
        html_content = file.read()
    
    # Display the HTML content
    display(HTML(html_content))
else:
    print(f"File not found: {file_path}")
    print("Please check if the path is correct and the file exists.")
```
